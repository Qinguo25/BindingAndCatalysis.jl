```julia
@syms x y z a b

# 2. Define the two rules for our two-stage approach

# STAGE 1 RULE: Moves a coefficient `c` to become an exponent for `b`.
coeff_rule = @rule ~c * log(~b) => log((~b)^(~c))

# STAGE 2 RULE: Adds two plain `log` terms together.
adder_rule = @rule log(~x) + log(~y) => log((~x) * (~y))


# 3. Create a rewriter for each stage
# The Fixpoint wrapper ensures each rule is applied until no more changes can be made.
coeff_rewriter = Fixpoint(Postwalk(coeff_rule))
adder_rewriter = Fixpoint(Postwalk(adder_rule))


# 4. Define expressions with one and four terms to test the solution
expr_one_term = 2log(x)
expr_four_terms = 2log(x) + a*log(y) + log(z)

# --- Test Case 1: A single term ---
println("--- Single Term Example ---")
println("Original: ", expr_one_term)

# First, apply the coefficient rule.
stage1_result = coeff_rule(expr_one_term)
println("After Stage 1 (coeffs): ", stage1_result)
# After Stage 1 (coeffs): log(x^2)

# The adder rule won't change anything, but we apply it for consistency.
final_result1 = adder_rewriter(stage1_result)
println("After Stage 2 (adding): ", final_result1)
# After Stage 2 (adding): log(x^2)

println("\n---------------------------\n")

# --- Test Case 2: Four terms ---
println("--- Four Term Example ---")
println("Original: ", expr_four_terms)

# Apply the two rewriters in sequence
stage1_result = coeff_rewriter(expr_four_terms)
println("After Stage 1 (coeffs): ", stage1_result)
# After Stage 1 (coeffs): log(z) + log(5^b) + log(x^2) + log(y^a)

final_result2 = adder_rewriter(stage1_result)
println("After Stage 2 (adding): ", final_result2)
# After Stage 2 (adding): log(z*(5^b)*(x^2)*(y^a))
```
